```python
import networkx as nx
import matplotlib.pyplot as plt
import snman

G = nx.Graph()
G.add_nodes_from([1,2,3])

G.add_node(1,ipos=(0,0))
G.add_node(2,ipos=(5,0))
G.add_node(3,ipos=(5,5))
G.add_node(4,ipos=(0,5))

G.add_edge(1,2,length=20,weight=1)
G.add_edge(2,3,length=20,weight=1)
G.add_edge(3,4,length=10,weight=1)
G.add_edge(4,1,length=20,weight=1)

G.add_edge(1,3,length=20,weight=1)
G.add_edge(2,4,length=20,weight=1)

a = nx.get_edge_attributes(G,'length')
b = {}
for key, value in a.items():
    u, v = key
    if not u in b:
        b[u] = {}
    b[u][v] = value

pos = nx.kamada_kawai_layout(G, dist=b, pos=nx.get_node_attributes(G,'ipos'), weight='weight')

fig = plt.figure(1, figsize=(5,5))
plt.xlim(-1.1,1.1)
plt.ylim(-1.1,1.1)
nx.draw_networkx(G, pos=pos)
plt.show()
```
